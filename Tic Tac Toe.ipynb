```python
#min-max algo
from copy import copy,deepcopy
humanTurn = True
turnCounts = 0;

#win check 
def check_win(win):
  for x in range(3):
    if(win[x][0] == win[x][1] and win[x][0] == win[x][2] and win[x][0] == 2):
      return 2
  for x in range(3):
    if(win[0][x] == win[1][x] and win[0][x] == win[2][x] and win[0][x] == 2):
      return 2

  if(win[0][0] == win[1][1] and win[0][0] == win[2][2] and win[0] == 2):
      return 2
  if(win[0][2] == win[1][1] and win[0][2] == win[2][0] and win[0][2] == 2):
      return 2

  for x in range(3):
    if(win[x][0] == win[x][1] and win[x][0] == win[x][2] and win[x][0] == 1):
      return 1
  for x in range(3):
    if(win[0][x] == win[1][x] and win[0][x] == win[2][x] and win[0][x] == 1):
      return 1
  if(win[0][0] == win[1][1] and win[0][0] == win[2][2] and win[0][0] == 1):
      return 1
  if(win[0][2] == win[1][1] and win[0][2] == win[2][0] and win[0][2] == 1):
      return 1
  
  check = 0 #0 check
  for i in range(3):
    for j in range(3):
      if(win[i][j] == 0):
        check = 1
        break
  if(check == 0):
    return 0
  else:
    return 3

def utility(board, depth):
  valid = check_win(board)
  if(valid == 1):
    return 1/depth
  elif(valid == 2):
    return -1
  elif(valid == 0):
    return 0

def max_algo(board, depth):
  valid = check_win(board)
  if(valid == 1 or valid == 2 or valid == 0):
    return utility(board, depth)
  util = -9999 #large neg number
  for i in range(3):
    for j in range(3):
      if(board[i][j] == 0):
        board1 = deepcopy(board)
        board1[i][j] = 1
        util = max(util, min_algo(deepcopy(board1), depth+1))
  return util      

def min_algo(board, depth):
  valid = check_win(board)
  if(valid == 1 or valid == 2 or valid == 0):
    return utility(board, depth)
  util = +9999 #large positive number
  for i in range(3):
    for j in range(3):
      if(board[i][j] == 0):
        board1 = deepcopy(board)
        board1[i][j] = 2
        util = min(util, max_algo(deepcopy(board1), depth+1))
  return util  

def choose_PC_move(mat):
  a = []
  for i in range(3):
    for j in range(3):
      if(mat[i][j] == 0):
        board1= deepcopy(mat)
        board1[i][j] = 1
        util = min_algo(deepcopy(board1), 1)
        
        a.append(PosTrack(i,j,util))
  max1 = -9999
  max2 = 0
  max3 = 0
  for x in a:
    if(max1 < x.val):
      max1 = x.val
      max2 = x.x
      max3 = x.y
  
  return max2, max3  

class PosTrack:
  def __init__(self,x,y,val):
    self.x = x
    self.y = y
    self.val = val

def inputCheck(x,y,board):
  if(x > 2 or y > 2 or x < 0 or y < 0): 
    return False
  if(board[x][y] != 0): 
    return False
  return True


def printBoard(board):
  for i in range(3):
    for j in range(3):
      if(board[i][j] == 2):
        print("X", end =" ")
      elif(board[i][j] == 1):
        print("O", end =" ")
      elif(board[i][j] == 0):
        print("*", end =" ")
    print()


# main
tictactoe_matrix = [[0 for x in range(3)] for y in range(3)]
printBoard(tictactoe_matrix)
while(True):
  axis_x = int(input("Enter row:"))           # player select position
  axis_y = int(input("Enter colomn:"))     
  if(not inputCheck(axis_x,axis_y,tictactoe_matrix)):
    continue
  tictactoe_matrix[axis_x][axis_y] = 2;
  turnCounts = turnCounts + 1

  print()
  print("Player's turn-")
  printBoard(tictactoe_matrix)
  
  valid = check_win(deepcopy(tictactoe_matrix))
  if(valid == 2):                               # player win check                            # 
    print()                                   
    print()                                   
    print("Game Over")                        
    print("Player wins")                      
    break

  if(turnCounts == 9):                        # draw check
    print()
    print()
    print("Game Over")  
    print("Draw The Game")
    break;


  humanTurn = False
  
  copy = deepcopy(tictactoe_matrix)
  m,n = choose_PC_move(copy) 
  tictactoe_matrix[m][n] = 1;
  turnCounts = turnCounts + 1

  print()
  print("Computer's turn-")
  printBoard(tictactoe_matrix)
  valid = check_win(deepcopy(tictactoe_matrix))

  if(valid == 1):                               # computer win check
    if(not humanTurn):                        
      print()                                 
      print()                                 
      print("Game Over")                      
      print("Computer wins")                  
      break
  if(turnCounts > 9):                        # draw check
    print()
    print()
    print("Game Over")  
    print("Draw The Game")
    break;
```
